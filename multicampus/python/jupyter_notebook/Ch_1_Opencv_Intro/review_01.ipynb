```python
# glob을 사용하면 한번에 파일 경로 및 이름을 합쳐서 생성해 준다.
img_lists = glob.glob('./fig/maple_story/*.*')
img_lists
```

```python
cv2.namedWindow('images', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('images', cv2.WND_PROP_FULLSCREEN,
                     cv2.WINDOW_FULLSCREEN)

idx = 0
while True:
    img = cv2.imread(img_lists[idx])
    
    cv2.imshow('images', img)
    
    idx += 1
    
    if cv2.waitKey(3000) == ord('q'):
        break
        
    if idx == len(img_lists):
        idx = 0
cv2.destroyAllWindows()    
```
