## Unit 15. pandas로 인구 구조 분석하기

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./data/age.csv', encoding='cp949', index_col=0)
df.head()
```

```python
df = df.div(df['총인구수'], axis=0) # 총인구수로 나누어줌(-> 데이터를 비율로 변환)
del df['총인구수'], df['연령구간인구수'] # 필요없는 데이터 삭제
df
```

```python
name =  input('원하는 지역의 이름을 입력해주세요 : ')
a = df.index.str.contains(name)  # 인덱스 문자열에서 입력값을 포함하는 행 찾기
df2 = df[a]
df2
```

```python
plt.rc('font', family='Malgun Gothic')
df2.T.plot()
plt.show
```

## <신도림동의 인구구조와 가장 비슷한 인구구조를 가진 지역 시각화>

### 1) 인구구조가 가장 비슷하다 => 인구비율의 "차이가 최소값"

```python
x = df.sub(df2.iloc[0], axis=1)
x
```

### 2) 음수값을 보정해줘야 함 => 제곱

```python
y = np.power(x,2)
y
```

### 3) 총 인구비율 => 모든 연령을 더함(모든 열을 더해줌)

```python
z = y.sum(axis=1)
z
```

### 4) 최솟값 정렬 => 오름차순으로 정렬

```python
i = z.sort_values().index[:5]
i
```

### 5) 시각화

```python
df.loc[i].T.plot()
plt.show()
```

###### ※ 만약 음수 보정을 하지 않는 다면?

```python
# 연령대별 인구비율 차이를 음수 보정하지 않고 총인구비율을 구함
z1 = x.sum(axis=1)
z1
```

```python
# 최솟값 순으로 정렬하여 5개 지역 추출. -> 앞과는 전혀 다른 결과값이 도출됨
i1 = z1.sort_values().index[:5]
i1
```

```python
df.loc[i1].T.plot()  
plt.show()
```
