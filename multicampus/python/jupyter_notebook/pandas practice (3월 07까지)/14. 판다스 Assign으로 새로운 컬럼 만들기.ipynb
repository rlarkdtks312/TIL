## assign으로 새로운 컬럼 만들기

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'A': range(1,11), 'B': np.random.randn(10)})
df
```

```python
df.assign?
```

```python
# df 객체에 접근해서 A 속성을 사용하여 각 요소에 자연로그를 취하게 함
df.assign(ln_A = lambda x: np.log(x.A))
```

```python
df.assign(ln_A = lambda x: np.log(x.A)).head()
```

```python
df['ln_A'] = np.log(df.A)
df
```
