##  2. 깔끔한 데이터 만들기(melt, pivot)

```python
df = pd.DataFrame({'A' : {0:'a', 1:'b', 2:'c'},
                   'B' : {0: 1, 1: 3, 2: 5},
                   'C' : {0: 2, 1: 4, 2: 6}})
df
```

```python
pd.melt(df, id_vars=['A'], value_vars=['B'])
```

```python
df2 = pd.melt(df, id_vars=['A'], value_vars=['B','C'])
df2
```

- method chaining

```python
df3 = pd.melt(df, value_vars=['A', 'B','C']).rename(columns={
    'variable': 'var',
    'value' : 'val'
})
df3
```
