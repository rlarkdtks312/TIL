## 멜론, 지니, 벅스 크롤링하기

```python
ser = Service('./chromedriver.exe')
driver = webdriver.Chrome(service = ser)
```

```python
url = 'https://www.melon.com/chart/index.htm'
driver.get(url)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
songs = soup.select('tbody > tr')
len(songs)
```

```python
# 순위, 제목, 가수

songs_order = soup.select('tbody > tr span.rank')
songs_title = soup.select('tbody > tr div.ellipsis.rank01')
songs_singer = soup.select('tbody > tr span.checkEllipsis')


list = [[],[],[]]


for i in range(len(songs_order)):
    list[0].append(songs_order[i].text)
    list[1].append(songs_title[i].text.replace('\n',''))
    list[2].append(songs_singer[i].text)
    
for i in range(100):
    print("{}위 곡은 {}입니다 {}님 축하드립니다.".format(list[0][i],list[1][i],list[2][i]))
```

```python
# DataFrame으로 만들기 위해 2차원 배열로 만들기

song_data = []

for i in range(len(songs_order)):
    song_data.append(['Melon', songs_order[i].text, songs_title[i].text.replace('\n',''), songs_singer[i].text])
    
print(song_data)
```

```python
# DateFrame으로 만들기

columns = ['서비스', '순위', '타이틀', '가수']
pd_data = pd.DataFrame(song_data, 
                       columns= columns)
pd_data
```

```python
# 엑셀로 저장하기

pd_data.to_excel('./files/melon_class.xlsx', index=False)
```

- 벅스 크롤링하기

```python
ser = Service('./chromedriver.exe')
driver = webdriver.Chrome(service = ser)
url = 'https://music.bugs.co.kr/chart'

driver.get(url)
```

```python
html = driver.page_source

soup = BeautifulSoup(html, 'html.parser')

soup
```

```python
orders = soup.select('tbody > tr td > div.ranking strong')
titles = soup.select('tbody > tr th a')
singers = soup.select('tbody > tr td > p.artist > a')

print('전')
print(len(orders))
print(len(titles))
print(len(singers), '\n')

# singers 8개의 추가값이 있어서 if 문으로 제거해 주었다.
for idx, word in enumerate(singers):
    if '\n' in word.text:
        singers.remove(singers[idx])

print('후')        
print(len(orders))
print(len(titles))
print(len(singers))
```

```python
bugs_data = []

for i in range(len(orders)) :
    bugs_data.append(['bugs', orders[i].text, titles[i].text, singers[i].text])
bugs_data
```

```python
columns = ['서비스', '순위', '타이틀', '가수']
pd_data_bugs = pd.DataFrame(bugs_data, 
                            columns=columns)
pd_data_bugs
```

```python
pd_data_bugs.to_excel('./files/bugs_class.xlsx', index = False)
```

- 지니 크롤링

```python
ser = Service('./chromedriver.exe')
driver = webdriver.Chrome(service = ser)
url = 'https://www.genie.co.kr/chart/top200'

driver.get(url)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
titles = soup.select('tbody > tr a.title.ellipsis')
singers = soup.select('tbody > tr a.artist.ellipsis')
temp = soup.select('tbody > tr')

print(type(soup))
print(type(titles))
print(len(titles))
print(len(singers))
```

```python
genie_data = []
order = 0
for i in range(len(titles)) :
    order +=1
    genie_data.append(['genie', order, titles[i].text.strip(), singers[i].text.strip()])
    
genie_data
```

```python
pd_data_genie = pd.DataFrame(genie_data, 
                             columns = ['서비스', '순위', '타이틀', '가수'])
pd_data_genie
```

```python
pd_data_genie.to_excel('./files/genie_class.xlsx', index=False)
```
