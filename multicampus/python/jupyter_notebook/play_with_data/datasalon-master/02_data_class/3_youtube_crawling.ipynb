## 유튜브 크롤링하기

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import time

import pandas as pd
import numpy as np
```

```python
## [음악/댄스, 가수명, 구독자수, 뷰수, 동영상수]
ser = Service('./chromedriver.exe')
driver = webdriver.Chrome(service=ser)
url = 'http://youtube-rank.com/board/bbs/board.php?bo_table=youtube'

driver.get(url)
```

```python
html = driver.page_source
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
category = soup.select('tbody > tr h1 > p')
names = soup.select('tbody > tr h1 > a')
subscriber = soup.select('tbody > tr > td.subscriber_cnt')
view = soup.select('tbody > tr > td.view_cnt')
video = soup.select('tbody > tr > td.video_cnt')

print(len(category))
print(len(names))
print(len(subscriber))
print(len(view))
print(len(video))
```

```python
data_youtube = []

for i in range(len(category)) :
    data_youtube.append([category[i].text.strip(),
                        names[i].text.strip(),
                        subscriber[i].text,
                        view[i].text,
                        video[i].text])
data_youtube
```

```python
columns = ['카테고리', '채널명', '구독자수', '총 플레이 수', '동영상 수']
pd_data_youtube = pd.DataFrame(data_youtube, columns=columns)
pd_data_youtube
```

```python
pd_data_youtube.to_excel('./files/youtube_class.xlsx', index = False)
```
