## 위와 같은 데이터 전처리 과정을 함수로 만들기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def create_kto_data(yy, mm) :
    # 1. 불러올 엑셀 파일 경로 지정
    file_path = './Tuorists_data/kto_{}{}.xlsx'.format(yy,mm)
    
    df = pd.read_excel(file_path, header = 1, usecols = 'A:G', skipfooter = 4)
    
    # 2. 기준년월 추가
    df['기준년월'] = '{}-{}'.format(yy,mm)
    
    # 3. 국적이 존재하는 데이터 값만 뽑아낸다.
    continents_list = ['아시아주', '미주', '구주', '대양주', '아프리카주',
                  '기타대륙', '교포소계']
    condition  = df['국적'].isin(continents_list)
    df_country = df[condition == False].reset_index(drop=True)
    
    # 4. 대륙 컬럼 추가하기
    continents = ['아시아']*25 + ['아메리카']*5 + ['유럽']*23 + ['오세아니아']*3 + ['아프리카']*2 + ['기타대륙'] + ['교포']
    df_country['대륙'] = continents
    
    # 5. 국가별 '관광객비율(%)' 컬럼 추가
    df_country['관광객비율(%)'] = round(df_country['관광']*100 / df_country['계'], 1)
    
    # 6. '전체비율(%)' 컬럼 추가
    tourist_sum = df_country['관광'].sum()
    df_country['전체비율(%)'] = round(df_country['관광']*100 / tourist_sum, 1)
    
    # 7. 결과 출력
    return df_country
```

```python
## 함수가 잘 실행되는 것을 확인 할 수 있다.
create_kto_data(2019, '02').head()
```

### 만든 함수를 사용하여 2010 ~ 2021.5월 까지의 데이터를 받기

```python
# zfill 빈칸앞에 0로 채워줌
# ignore_index = True 인덱스 새로 만들어줌

df_total = pd.DataFrame()

for yy in range(2010, 2021): 
    for mm in range(1, 13): 
        try:
            temp = create_kto_data(yy,str(mm).zfill(2))
            df_total = df_total.append(temp, ignore_index=True)
        except:
            pass
    
df_total.head(20)
```

### 국적별로 저장하기

```python
def country_save(country) :
    file_path = './Tuorists_data/my_total.xlsx'
    df = pd.read_excel(file_path, header = 0, usecols = 'A:K')
    
    condition = df['국적'] == country
    
    df = df[condition].sort_values(by = '관광객비율(%)', ascending=False)

    return df.reset_index(drop=True)
```

```python
for i in range(len(df_total['국적'].unique())):
    print(df_total['국적'].unique()[i])
```

```python
for i in range(len(df_total['국적'].unique())) :
    country = df_total['국적'].unique()[i]
    temp = country_save(country)
    temp.to_excel('./Tuorists_data/my_[국적별 관광객 데이터] {}.xlsx'.format(country))
```
