```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import rc, font_manager

import seaborn
```

```python
path = 'c:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname = path).get_name()

rc('font', family = font_name)
```

```python
df = pd.read_excel('./Tuorists_data/my_total.xlsx')
df
```

```python
df_filter = df[df['국적'] == '중국']
df_filter.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_filter['기준년월'], df_filter['관광'])
plt.title('중국인 관광객 추세')
plt.xlabel('기준년월')
plt.ylabel('관광객수')
plt.xticks(['2010-01', '2011-01', '2012-01', '2013-01', '2014-01',
            '2015-01', '2016-01', '2017-01', '2018-01', '2019-01', '2020-01'])
plt.show()
```

```python
cntry_list = ['중국', '일본', '대만', '미국', '홍콩']
```

```python
for cntry in cntry_list:
    condition = df['국적'] == cntry
    df_filter = df[condition]
    
    plt.figure(figsize=(12,4))
    
    plt.plot(df_filter['기준년월'], df_filter['관광'])
    plt.title('{} 관광객 추세'.format(cntry))
    plt.xlabel('기준년월')
    plt.ylabel('관광객수')
    plt.xticks(['2010-01', '2011-01', '2012-01', '2013-01', '2014-01',
                '2015-01', '2016-01', '2017-01', '2018-01', '2019-01', '2020-01'])
    plt.show()
```

### 히트맵 그래프 그리기
- 가로방향은 월을 그리고 세로는 년도를 나타내어 보자!

```python
df.head()
```

```python
df['년도'] = df['기준년월'].str.slice(0,4)
df['월'] = df['기준년월'].str.slice(5,7)

df.head()
```

```python
condition = df['국적'] == '중국'

df_filter = df[condition]
df_filter.head()
```

```python
df_pivot = df_filter.pivot_table(values='관광',
                                index = '년도',
                                columns = '월')

df_pivot
```

```python
import seaborn as sns
```

```python
plt.figure(figsize = (16, 10))
sns.heatmap(df_pivot, annot = True, fmt = '.0f', cmap = 'rocket_r')
plt.show()
```

- seaborn 공부

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights_pivot = flights.pivot_table(values = 'passengers',
                                   index = 'year',
                                   columns = 'month')
flights_pivot
```

### plt로 heatmap 그리기 및 seaborn 공부

```python
# plt로 heatmap 그리기
plt.pcolor(flights_pivot, cmap='winter')
plt.xticks(np.arange(0.5, len(flights_pivot.columns)))
plt.yticks(np.arange(0.5, len(flights_pivot.index)))
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.heatmap(flights_pivot,
           annot = True, 
           fmt = 'd',
           cmap = 'spring',
           annot_kws = {'size': 10, 'color':'black'})
plt.title('HeatMap', fontsize = 10)
```
