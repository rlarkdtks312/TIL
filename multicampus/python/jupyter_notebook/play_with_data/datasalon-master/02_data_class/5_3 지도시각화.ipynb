```python
import requests
import privacy as pr # API 키를 저장해둔 파일

searching = 'Jeju'
url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(searching)

headers = {
    "Authorization": 'KakaoAK {}'.format(pr.K_API) 
}

places = requests.get(url, headers = headers).json()['documents']
print(places[0]['place_name'])
print('경도 =', places[0]['x'])
print('위도 =', places[0]['y'])
```

```python
def find_places(searching):
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(searching)
    
    headers = {
    "Authorization": 'KakaoAK {}'.format(pr.K_API) 
    }
    
    places = requests.get(url, headers = headers).json()['documents']
    
    place = places[0]
    
    name = place['place_name']
    x = place['x']
    y = place['y']
    
    data = [name, x, y, searching]
    
    return data
```

```python
data = find_places('제주공항')
data
```

```python
from tqdm.notebook import tqdm
```

```python
import time

loca_inform = []

for location in tqdm(locations[:200]):
    try:
        data = find_places(location)
        loca_inform.append(data)
        time.sleep(1)
    except:
        pass
loca_inform
```

```python
len(loca_inform)
```

```python
location_inform_df = pd.DataFrame(loca_inform)
location_inform_df.columns = ['name_official', 'x', 'y', 'name']
location_inform_df.head()
location_inform_df.to_excel('./my_files/3_locations_class.xlsx', index=False)
```

```python
locations_counts_df = pd.read_excel('./5_Jeju_Hotplace/files/3_location_counts.xlsx', index_col = 0)
locations_inform_df = pd.read_excel('./5_Jeju_Hotplace/files/3_locations.xlsx')
```

```python
location_counts_df.head()
```

```python
location_inform_df.head()
```

```python
# inner는 교집합이 있는 것만 뽑아 낸다는 의미이다.
location_data = pd.merge(locations_inform_df, locations_counts_df, 
                         how = 'inner', left_on = 'name_official', right_index=True)

location_data.head()
```

```python
location_data['name_official'].value_counts()
```

```python
location_data = location_data.pivot_table(values='place',
                             index = ['name_official', '경도', '위도'],
                             aggfunc='sum')
location_data.sort_values(by = 'place', ascending=False)
```

### 지도를 그려주는 모듈
- 맵객체를 만들어 주고 쌓아 올리는 방식이다.
- fa : font awesome 에서 필요에 따른 여러 아이콘을 받아 사용할 수 있다.
- popup에 유튜브 링크를 넣을 수 있다.

```python
import folium
```

```python
# 기본좌표
latitude = 37.394946
longitude = 127.111104

m = folium.Map(location= [latitude, longitude],
              width = 700,
              hight = 500,
              zoom_start = 15)
folium.Marker(location = [latitude, longitude],
             popup = '<iframe width="896" height="504" src="https://www.youtube.com/embed/yJjC6NGWyMg" title="YouTube video player" frameborder="0" allow="accelerometer; autoplay; clipboard-write; encrypted-media; gyroscope; picture-in-picture" allowfullscreen></iframe>',
             tooltip = '백크',
             icon = folium.Icon(color='red',
                               icon_color='white',
                               icon = 'ambulance',
                               prefix = 'fa')).add_to(m)
folium.CircleMarker(location=[latitude, longitude],
                   color = 'tomato',
                   radius = 50,
                   tooltip = '활동반경').add_to(m)
m
```

```python
import folium

# 지도객체 만들고 시작점 시작 배율 설정
Mt_Hanla =[33.362500, 126.533694]
map_jeju = folium.Map(location = Mt_Hanla, zoom_start = 11)

# 지도 형식 변경
folium.TileLayer().add_to(map_jeju)

# 반복문 형식으로 여러 위경도 및 원형마크 표시
for i in range(len(location_data)):
    name = location_data['name_official'][i]
    size = int(location_data['place'][i]) * 2
    lat = location_data['위도'][i]
    long = location_data['경도'][i]
    folium.CircleMarker(location=[lat, long], radius=size, color = 'red', popup=name).add_to(map_jeju)

map_jeju
```

### tilelayer를 여러개 쌓아서 지도 생성하기

```python
from folium.plugins import MarkerCluster
```

```python
locations = []
names = []

for i in range(len(location_data)):
    data = location_data.iloc[i]
    locations.append([data['위도'], data['경도']])
    names.append(data['name_official'])

# 지도객체 만들고 시작점 시작 배율 설정
Mt_Hanla =[33.362500, 126.533694]
map_jeju2 = folium.Map(location = Mt_Hanla, zoom_start = 10)

tiles = ["Stamen Toner", "Stamen Watercolor",
         "CartoDB positron", "CartoDB dark_matter", 'MapQuest Open Aerial']

for tile in tiles:
    folium.TileLayer(tile).add_to(map_jeju2)
    
marker_cluster = MarkerCluster(
    locations=locations, popups=names,
    name='Jeju',
    overlay=True,
    control=True,

)

marker_cluster.add_to(map_jeju2)
folium.LayerControl().add_to(map_jeju2)
map_jeju2
```
