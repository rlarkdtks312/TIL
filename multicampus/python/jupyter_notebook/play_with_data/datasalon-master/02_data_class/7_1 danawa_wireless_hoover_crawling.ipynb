```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By

import time
from tqdm import tqdm_notebook

import pandas as pd
import numpy as np
```

```python
ser = Service('./chromedriver.exe')
driver = webdriver.Chrome(service=ser)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
# try: except: 구문으로 같은 태그내에 들어있는 광고를 제거 할 수 있었다.
wireless_hoover = []

for i in tqdm_notebook(range(len(product_item))):
    try:
        name = product_item[i].select('div.prod_info p.prod_name')[0].text.strip()
        spec = product_item[i].select('div.prod_info dl.prod_spec_set')[0].text.replace('\n','').replace('\t','')
        cost = product_item[i].select('div.prod_pricelist li.rank_one p.price_sect a')[0].text.replace('원', '').replace(',', '')

        temp = [name, spec, cost]
        wireless_hoover.append(temp)
    except:
        pass

print(len(wireless_hoover))
wireless_hoover
```

```python
def get_prod_items(product_item) :
    prod_data = []
    for i in tqdm_notebook(range(len(product_item))):
        try:
            name = product_item[i].select('div.prod_info p.prod_name')[0].text.strip()
            spec = product_item[i].select('div.prod_info dl.prod_spec_set')[0].text.replace('\n','').replace('\t','')
            cost = product_item[i].select('div.prod_pricelist li.rank_one p.price_sect a')[0].text.replace('원', '').replace(',', '')

            temp = [name, spec, cost]
            prod_data.append(temp)
        except:
            pass
    return prod_data
```

```python
def get_search_page_url(keyword, page):
    url = 'http://search.danawa.com/dsearch.php?query={}&originalQuery={}&previousKeyword={}&volumeType=allvs&page={}&limit=40&sort=saveDESC&list=list&boost=true&addDelivery=N&recommendedSort=Y&defaultUICategoryCode=102207&defaultPhysicsCategoryCode=72%7C80%7C81%7C0&defaultVmTab=2606&defaultVaTab=390042&tab=goods'.format(keyword, keyword, keyword, page)
    return url
```

```python
get_search_page_url('무선청소기', 2)
```

```python
keyword = '무선청소기'
total_page = 10
prod_data_total = []

for page in tqdm_notebook(range(1, total_page+1)) :
    url = get_search_page_url(keyword, page)
    
    driver.get(url)
    
    time.sleep(3)
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    product_item = soup.select('div.main_prodlist ul.product_list > li.prod_item')
    
    prod_item_list = get_prod_items(product_item)
    prod_data_total.append(prod_item_list)
```

```python
total = []
for temp in prod_data_total:
    total += temp
prod_data_total = total
prod_data_total
```

```python
data = pd.DataFrame(prod_data_total)

data.columns = ['상품명', '스펙 목록', '가격']
data.to_excel('./my_files/danawa_crawling_result_class.xlsx', index =False)
```
