```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./titanic.csv')
df
```

```python
# replace
# male -> 0 , female -> 1
df.Sex = df.Sex.replace({'male':0, 'female':1})
```

```python
df
```

```python
# 수치형 자료 변환
# 정규화(normalization)

# 1) 정규화 함수 구현

def normal(data):
    max_v = max(data)
    min_v = min(data)
    data = (data - min_v) / (max_v - min_v)
    return data

df.Fare.head()
```

```python
Fare_N = normal(df.Fare)
df.Fare = Fare_N
print(Fare_N)
```

```python
# 수치형 자료 변환
# 표준화(Standardization)

# 2) 표준화 함수 구현

def standard(data):
    data = (data - data.mean()) / data.std()
    return data
```

```python
Fare_SD = standard(df.Fare)
df.Fare = Fare_SD
print(Fare_SD)
```
