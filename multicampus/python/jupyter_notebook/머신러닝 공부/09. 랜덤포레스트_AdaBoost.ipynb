```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import tree, ensemble
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
```

```python
# 데이터 로딩
# 중산층이냐 아니냐를 판펼
pd.read_csv('./middle.csv - middle.csv.csv')
```

```python
middle = pd.read_csv('./middle.csv - middle.csv.csv')
```

```python
x = middle.iloc[:,:-1].values
y = middle.iloc[:,-1].values
```

```python
x.shape, y.shape
```

```python
# 정규화
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
```

```python
x_regu = x
x_regu
```

```python
# 랜덤포레스트
classifier = ensemble.RandomForestClassifier(n_estimators=20, max_depth=3, criterion='gini')
```

```python
classifier.fit(x_regu, y)
```

```python
x_regu, type(x_regu), y, type(y)
```

```python
# 실제값

expected = y
expected
```

```python
# 예측값
predicted = classifier.predict(x_regu)
predicted
```

```python
print("Accuracy:\n",
     accuracy_score(expected, predicted))
print()
print('Confusion_Matrix:\n',
     confusion_matrix(expected, predicted))
```

## 에이다 부스트 AdaBoost

```python
# base 를 decision tree로 하겠다.
classfier2 = ensemble.AdaBoostClassifier(base_estimator=tree.DecisionTreeClassifier(max_depth=3), n_estimators=20)
```

```python
classfier2.fit(x_regu, y)
```

```python
# 분류기 성능 확인

expected2 = y
predicted2 = classfier2.predict(x_regu)

print(predicted2)
```

```python
print("Accuracy:\n",
     accuracy_score(expected2, predicted2))
print()
print('Confusion_Matrix:\n',
     confusion_matrix(expected2, predicted2))
```
