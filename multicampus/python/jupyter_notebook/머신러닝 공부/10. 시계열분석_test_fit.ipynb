```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# 위 그래프의 acf, pacf 곡선 그리기 
# 사용할 데이터는 data
plot_acf(data)
plot_pacf(data)
plt.show()

# 벗어난 애들이 너무 많다 한번 더 차분해야 겠다.
```

```python
diff_1 = data.diff(periods=1).iloc[1:]
diff_1.plot()
```

```python
plot_acf(diff_1)
plot_pacf(diff_1)
plt.show()
```
