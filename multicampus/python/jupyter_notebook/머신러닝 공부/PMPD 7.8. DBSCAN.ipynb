```python
# 지도 위치 표시

mschool_map = folium.Map(location=[37.55,126.98], tiles='Stamen Terrain', 
                        zoom_start=12)
```

```python
# 중학교 위치 정보를 circle marker로 표시

for name, lat, log in zip(df.학교명, df.위도, df.경도):
    folium.CircleMarker([lat, log],
                       radius=5, # 원의 반지름
                       color='brown',
                       fill=True,
                       fill_color='coral',
                       fill_opacity=0.7,
                       popup = name,
                       ).add_to(mschool_map)
mschool_map
```

```python
mschool_map.save('./seoul_mschool.html')
```

```python
# 데이터 전처리

# 원 핫 인코딩 (더미 변수 생성)

from sklearn.preprocessing import LabelEncoder, OneHotEncoder

label_encoder = LabelEncoder()
oh = OneHotEncoder()
```

```python
df.info()
```

```python
onehot_location = label_encoder.fit_transform(df['지역'])
onehot_code = label_encoder.fit_transform(df['코드'])
onehot_type = label_encoder.fit_transform(df['유형'])
onehot_day = label_encoder.fit_transform(df['주야'])
```

```python
df['location'] = onehot_location
df['code'] = onehot_code
df['type'] = onehot_type
df['day'] = onehot_day
```

```python
df.head()
```

```python
# DBSCAN 군집 모형

from sklearn import cluster

# 분석에 사용할 속성 선택
columns_list = [9, 10, 13]
X = df.iloc[:, columns_list]
print(X[:5])
print('\n')
```

```python
from sklearn.preprocessing import StandardScaler

StandardScaler().fit_transform(X)
```

```python
X = StandardScaler().fit_transform(X)
X
```

```python
# DBSCAN 모델 생성

db_model = cluster.DBSCAN(eps=0.2, min_samples=5)
```

```python
db_model.fit(X)
```

```python
# 결과 확인

db_model.labels_
```

```python
cluster_label = db_model.labels_
```

```python
# 예측 결과를 데이터 프레임에 추가

df['Cluster']=cluster_label
df.head()
```

```python
columns_list
```

```python
# 클러스터 값으로 그룹화하고, 그룹별로 내용 출력 (첫 5행만 출력)
grouped_cols = [0, 1, 3] + columns_list
grouped = df.groupby('Cluster')
for key, group in grouped:
    print('* key :', key)
    print('* number :', len(group))    
    print(group.iloc[:, grouped_cols].head())
    print('\n')
```

```python
# 그래프로 표현 - 시각화
colors = {-1:'gray', 0:'coral', 1:'blue', 2:'green', 3:'red', 4:'purple', 
          5:'orange', 6:'brown', 7:'brick', 8:'yellow', 9:'magenta', 10:'cyan'}

cluster_map = folium.Map(location=[37.55,126.98], tiles='Stamen Terrain', 
                        zoom_start=12)

for name, lat, lng, clus in zip(df.학교명, df.위도, df.경도, df.Cluster):  
    folium.CircleMarker([lat, lng],
                        radius=5,                   # 원의 반지름
                        color=colors[clus],         # 원의 둘레 색상
                        fill=True,
                        fill_color=colors[clus],    # 원을 채우는 색
                        fill_opacity=0.7,           # 투명도    
                        popup=name
    ).add_to(cluster_map)
```

```python
cluster_map
```

```python
# 지도를 html 파일로 저장하기
cluster_map.save('./seoul_mschool_2.html')
```

```python
# X2 데이터셋에 대하여 위의 과정을 반복(과학고, 외고국제고, 자사고 진학률 + 유형)
columns_list2 = [9, 10, 13, 22]
X2 = df.iloc[:, columns_list2]
print(X2[:5])
print('\n')
```
