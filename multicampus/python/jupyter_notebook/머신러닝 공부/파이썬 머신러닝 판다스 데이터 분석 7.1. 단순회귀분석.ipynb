```python
# -*- coding: utf-8 -*-

### 기본 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 문자열(object) >>> 숫자 
df.horsepower.replace('hp', np.nan, inplace=True)
# 행 하나 날리기
df = df.drop(0,axis=0)
df
```

```python
df.horsepower = df.horsepower.astype('float') # 실수형으로 변경
df.mpg = df.mpg.astype('float') 
df.cylinders = df.cylinders.astype('int')
df.displacement = df.displacement.astype('float')
df.weight = df.weight.astype('float')
df['model year'] = df['model year'].astype('float')
df.origin = df.origin.astype('int')
df
```

## [Step 3] 속성(feature 또는 variable) 선택

```python
ndf = df[['mpg','cylinders', 'horsepower', 'weight']]
ndf
```

```python
### 종속 변수 Y인 "연비(mpg)"와 다른 변수 간의 선형관계를 그래프(산점도)로 확인
# 산점도 그리기
ndf.plot(kind='scatter', x='weight', y='mpg', color='coral',s=10,figsize=(10,5))
```

```python
# seaborn
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
sns.regplot(x='weight', y='mpg', data=ndf, ax=ax1) # 회귀선 있음
sns.regplot(x='weight', y='mpg', data=ndf, ax=ax2, fit_reg=False) # 회귀선 없음
plt.show()
```

```python
# seaborn 연결 그래프(jointplot()) : 산점도, 히스토그램

sns.jointplot(x='weight', y='mpg', data=ndf)
sns.jointplot(x='weight', y='mpg', kind='reg', data=ndf) # 회귀선 있음
```

```python
# 작극 추천 
sns.pairplot(ndf)
plt.show()
```

## Step 4: 데이터셋 구분 - 훈련용(train data)/ 검증용(test data)1

```python
ndf[['weight']] # 데이터 프레임 형태로 반환
```

```python
ndf['weight'] # 시리즈 형태로 반환
```

```python
ndf['weight'].shape
```

```python
X = ndf[['weight']] # 독립변수
y = ndf['mpg'] # 종속변수
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
print(len(X_train))
print(len(X_test))
```

##  Step 5: 단순회귀분석 모형 - sklearn 사용


```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train) # fit:  train data로 학습 
```

```python
# 성능평가(score) >> 결정계수(r-제곱 : 설명력) 계산

r_2 = lr.score(X_test, y_test)
r_2
```

```python
lr.coef_ # 회귀식의 기울기
```

```python
lr.intercept_ # 회귀식의 y절편
```

```python
# 모형 전체 x 데이터 입력, y(실제값)과 y_hat(예측값) 비교
lr.predict(X)
```

```python
y_hat = lr.predict(X)
```

```python
plt.figure(figsize=(10,5))
ax1 = sns.kdeplot(y, label='y')
ax2 = sns.kdeplot(y_hat,label='y_hat', ax=ax1)
plt.show()
```
